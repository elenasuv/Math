Задание 1.2 Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def length_vector(vec):
    length_vector = np.sqrt(sum(i**2 for i in vec))
    return length_vector
```

```python
a = (0, 2, 3, 4)
length_vector(a)
```

Задание 3. Напишите код на Python, реализующий построение графиков:
окружности,	
эллипса,
гиперболы.


```python
#окружность
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
R = 2
a = (1, 2)
def circle(x, y, a, R):
    X, Y = np.meshgrid(x,y)
    circle = (X-a[0])**2 + (Y-a[1])**2 - R**2
    plt.contour(X,Y,circle,[0])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect('equal')
    plt.show()   
```

```python
circle(x, y, a, R)
```

```python
#эллипс
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
a = 2
b = 3
m = (0, 0)
def ellipse(x, y, a, b, m):
    X, Y = np.meshgrid(x,y)
    ellipse = (X-m[0])**2 / a**2 + (Y-m[1])**2 / b**2 - 1
    plt.contour(X,Y,ellipse,[0])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect('equal')
    plt.show()   
```

```python
ellipse(x, y, a, b, m)
```

```python
#гипербола
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
a = 2
b = 3
m = (0, 0)
def hyperbole(x, y, a, b, m):
    X, Y = np.meshgrid(x,y)
    ellipse = (X-m[0])**2 / a**2 - (Y-m[1])**2 / b**2 - 1
    plt.contour(X,Y,ellipse,[0])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect('equal')
    plt.show()   
```

```python
hyperbole(x, y, a, b, m)
```

Задание 5.1 Нарисуйте трехмерный график двух параллельных плоскостей

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)
Z1 = 4 - 5*X - 3*Y
Z2 = 300 - 5*X - 3*Y
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```

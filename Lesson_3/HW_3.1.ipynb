```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20, 20, 100)
y1 = 2*np.cos(x - 1) + 3
y2 = np.cos(x + 5) + 2
y3 = 4*np.cos(x - 7) + 10
plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")
plt.plot(x, y3, label="y3")
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-10, 10, 0, 20])
plt.legend(loc="upper right")
plt.show()
```

4.1 –†–µ—à–∏—Ç–µ —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π exp(x) + x‚àô(1 ‚Äì y) = 1  
y = x**2 ‚Äì 1

```python
x = np.linspace(-20, 20, 100)
y1 = x**2 - 1
y2 = (np.exp(x) + x - 1) / x
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2,30)
plt.xlim(-5,7)
plt.grid(True)
plt.legend(loc="upper right")
plt.show()
```

```python
from scipy.optimize import fsolve 
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x* (1 - y) - 1)
x1, y1 = fsolve(equations, (-2, 5))
print(x1, y1)
```

```python
x2, y2 = fsolve(equations, (5, 25))
print(x2, y2)
```

```python
x3, y3 = fsolve(equations, (3, 9))
print(x3, y3)
```

4.2 –†–µ—à–∏—Ç–µ —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∏ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ exp(x) + x‚àô(1 ‚Äì y) - 1 > 0 
y = x**2 ‚Äì 1

```python
# —Ä–µ—à–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ y = x**2 ‚Äì 1
def f(x):
    return (x**2 - 1)
x1 = fsolve(f, -2)
print(x1)
```

```python
x2 = fsolve(f, 2)
print(x2)
```

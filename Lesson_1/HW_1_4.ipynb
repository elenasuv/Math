```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
k1 = 1
k2 = 1.5
plt.plot(x, np.cos(k1*x), label="y = cos(x)")
plt.plot(x, np.cos(k2*x), label="y = cos(1.5x)")
plt.title("График функции y = cos(kx)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="lower right")
plt.axis([-10, 10, -1.5, 1.1])
plt.show()
```
